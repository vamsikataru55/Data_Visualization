```python
#Filling the null in Age Column
data["Age"]=data["Age"].fillna(data["Age"].median())
data
```

```python
#In cabin column having more than 50% null values so, we are droping the cabin column
data.drop(["Cabin"],axis=True)
```

```python
#In Fare column having one null value.so, we are filling with median value
data["Fare"]=data["Fare"].fillna(data["Fare"].median())
data["Fare"].isnull().sum()
```

```python
df=data[["Age","Fare"]]
df
```

```python
df.shape
```

```python
#Standard Scaling
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
scaler.fit(df)
df1=scaler.transform(df)
df1
```

```python
df_scaled=pd.DataFrame(df1,columns=df.columns)
df_scaled
```

```python
data1=pd.concat([data,df_scaled], axis=True)
data1
```

```python
data1.isnull().sum()
```

```python
data1=data1.dropna()
data1.isnull().sum()
```
