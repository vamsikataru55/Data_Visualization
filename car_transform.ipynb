```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("car_prices.csv")
data
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

for col in data.select_dtypes(include=['object']).columns:
    data[col] = encoder.fit_transform(data[col])
```

```python
data.info()
```

```python
from sklearn.preprocessing import PowerTransformer
power = PowerTransformer()

for col in data.select_dtypes(include=[np.int64,np.float64,np.int32]).columns:
    data[col] = power.fit_transform(data[col].values.reshape(-1,1))
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

features= data.select_dtypes(include=[np.int64,np.float64]).drop(['sellingprice'],axis=1)
target = data.select_dtypes(include=[np.int64,np.float64]).loc[:,'sellingprice']

x_train,x_test,y_train,y_test = train_test_split(features, target, test_size=0.2, random_state=214)
```
